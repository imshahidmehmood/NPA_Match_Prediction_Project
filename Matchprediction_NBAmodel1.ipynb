```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Import necessary libraries for model selection
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

# Initialize the models
logistic_regression = LogisticRegression()
random_forest = RandomForestClassifier()
gradient_boosting = GradientBoostingClassifier()
```

```python
# Train the models
logistic_regression.fit(X_train, y_train)
random_forest.fit(X_train, y_train)
gradient_boosting.fit(X_train, y_train)

# Evaluate the models
logistic_regression_score = logistic_regression.score(X_test, y_test)
random_forest_score = random_forest.score(X_test, y_test)
gradient_boosting_score = gradient_boosting.score(X_test, y_test)

print("Logistic Regression Accuracy:", logistic_regression_score)
print("Random Forest Accuracy:", random_forest_score)
print("Gradient Boosting Accuracy:", gradient_boosting_score)
```

```python
# Compare model performance
if logistic_regression_score > random_forest_score and logistic_regression_score > gradient_boosting_score:
    best_model = logistic_regression
    best_model_name = "Logistic Regression"
elif random_forest_score > logistic_regression_score and random_forest_score > gradient_boosting_score:
    best_model = random_forest
    best_model_name = "Random Forest"
else:
    best_model = gradient_boosting
    best_model_name = "Gradient Boosting"

print("Best Model:", best_model_name)
```
